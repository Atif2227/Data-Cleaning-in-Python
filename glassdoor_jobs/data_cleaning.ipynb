1. salary parsing 

```python
# Removing -1s
df = df[df['Salary Estimate'] != '-1']
```

```python
df.shape
```

we have removed 956-742 rows with Salary Estimate = -1

```python
# Parsing Salary Estimate by '('
df["Salary Estimate"] = df["Salary Estimate"].apply(lambda x: x.split('(')[0])
```

```python
# Removing K and $ from Salary Estimate
df["Salary Estimate"] = df["Salary Estimate"].apply(lambda x: x.replace('K','').replace('$',''))
df
```

```python
# Removing 'Per Hour' & 'Employer Provided Salary'
df["Salary Estimate"] = df["Salary Estimate"].apply(lambda x: 
x.replace('Per Hour','').replace('Employer Provided Salary',''))
df
```

```python
# Removing ':'
df["Salary Estimate"] = df["Salary Estimate"].apply(lambda x: 
x.replace(':',''))
df
```

```python
# Adding 'min_Salary' column
df["min_salary"] = df["Salary Estimate"].apply(lambda x: int(x.split('-')[0]))
```

```python
# Adding 'max_Salary' column
df["max_salary"] = df["Salary Estimate"].apply(lambda x: int(x.split('-')[1]))
```

```python
# Adding 'max_Salary' column
df["avg_salary"] = (df.min_salary + df.max_salary)/2
```

2. company name only

```python
df["Company Name"] = df.apply(lambda x: 
    x["Company Name"] if x["Rating"] < 0 
                        else x["Company Name"][:-3], axis = 1 
)
```

3. state field

```python
df["State"] = df["Location"].apply(lambda x: x.split(',') [1])
```

```python
# BTW, we can see how many jobs are there in each states and that will be equal to frequency of states.
df.State.value_counts()
```

4. age of company

```python
df.columns
```

```python
# Let's make a column to check if job Location and Headquarter are in same state or not.
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
```

```python
# Age of the company
from datetime import date
df["company age"] = df.Founded.apply(lambda x: x if x<1 else date.today().year-x )
```
