```python
df = pd.read_csv('glassdoor_jobs.csv')
df.head()
```

```python
df.shape
```

```python
# 1. salary parsing (Removing -1s)
df = df[df['Salary Estimate'] != '-1']
```

```python
df.shape
```

we have removed 956-742 rows with Salary Estimate = -1

```python
# Parsing Salary Estimate by '('
df["Salary Estimate"] = df["Salary Estimate"].apply(lambda x: x.split('(')[0])
```

```python
# Removing K and $ from Salary Estimate
df["Salary Estimate"] = df["Salary Estimate"].apply(lambda x: x.replace('K','').replace('$',''))
df
```

```python
# Removing 'Per Hour' & 'Employer Provided Salary'
df["Salary Estimate"] = df["Salary Estimate"].apply(lambda x: 
x.replace('Per Hour','').replace('Employer Provided Salary',''))
df
```

```python
# Removing ':'
df["Salary Estimate"] = df["Salary Estimate"].apply(lambda x: 
x.replace(':',''))
df
```
